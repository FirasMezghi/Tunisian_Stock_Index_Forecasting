```python
mylines = []                             # Declare an empty list named mylines.
with open ('/content/drive/MyDrive/dataset ml/stock_tunisia/histo_indice_2013.txt', 'rt') as myfile: # Open lorem.txt for reading text data.
    for myline in myfile:                # For each line, stored as myline,
        mylines.append(myline[:-2])           # add its contents to mylines.
print(mylines)
```

```python
mylines.pop(1)
print(mylines)
```

```python
for i in range(len(mylines)):
  mylines[i]=mylines[i].split(' ')
  j=0
  while(j <len(mylines[i])):
    if mylines[i][j]=='':
      mylines[i].pop(j)
    else:
      j+=1
 
print(mylines)
```
