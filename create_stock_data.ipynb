```python
df_15=pd.read_csv('/content/drive/MyDrive/dataset ml/stock_tunisia/stock_index_2015.csv')
df_15['LIB_INDICE']=list(df_15['LIB_IND'])
df_14=pd.read_csv('/content/drive/MyDrive/dataset ml/stock_tunisia/stock_index_2014.csv')
df_13=pd.read_csv('/content/drive/MyDrive/dataset ml/stock_tunisia/stock_index_2013.csv')
```

```python
def create_stock_data(df):
  stock_index=df['LIB_INDICE'].unique()
  df_stock=pd.DataFrame(df['SEANCE'].unique(),columns=['DATE'])
  for i in stock_index:
    data=df[df['LIB_INDICE']==i]
    if len(list(data['INDICE_JOUR']))==len(df_stock):
      df_stock[i]=list(data['INDICE_JOUR'])
  return df_stock
```

```python
data_18=create_stock_data(df_18)
data_17=create_stock_data(df_17)
data_16=create_stock_data(df_16)
data_15=create_stock_data(df_15)
data_13=create_stock_data(df_13)
data_14=create_stock_data(df_14)
data=pd.concat([data_13,data_14,data_15,data_16,data_17,data_18], axis=0)
data
```

```python
data.to_csv('index_stock_tunisia.csv',index=False)
```
